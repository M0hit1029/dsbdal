```python
import requests

url = "https://store.steampowered.com/appreviews/730?json=1"
response = requests.get(url)
data = response.json()

for review in data['reviews']:
    print(f"User: {review['author']['steamid']}")
    print(f"Review: {review['review']}")
    print(f"Rating: {'Recommended' if review['voted_up'] else 'Not Recommended'}")
    print("-" * 20)
```

```python
import requests
from bs4 import BeautifulSoup

url = "http://books.toscrape.com/"
soup = BeautifulSoup(requests.get(url).text, 'html.parser')

# Find all book containers
books = soup.find_all('article', class_='product_pod')

for book in books:
    name = book.h3.a['title']
    price = book.find('p', class_='price_color').text
    rating = book.p['class'][1] # e.g., "Three", "Four"
    print(f"Book: {name} | Price: {price} | Rating: {rating}")
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://webscraper.io/test-sites/e-commerce/static/computers/laptops"

def scrape_ecommerce():
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')

        products = soup.find_all('div', class_='thumbnail')

        print(f"{'Product Name':<25} | {'Price':<10} | {'Reviews':<12} | {'Rating'}")
        print("-" * 65)

        for product in products:
            name_tag = product.find('a', class_='title')
            name = name_tag['title'] if name_tag else "N/A"

            price_tag = product.find('span', itemprop='price')
            price = price_tag.text if price_tag else "N/A"

            desc_tag = product.find('p', class_='description')
            desc = desc_tag.text.strip() if desc_tag else "No description"

            review_tag = product.find('p', class_='review-count')
            review_count = review_tag.text.strip() if review_tag else "0 reviews"

            rating_tag = product.find('p', {'data-rating': True})
            rating = rating_tag['data-rating'] if rating_tag else "0"

            print(f"{name[:25]:<25} | {price:<10} | {review_count:<12} | {rating} Stars")
            print(f"Comment: {desc[:75]}...")
            print("-" * 65)

    except Exception as e:
        print(f"An error occurred: {e}")

scrape_ecommerce()
```
