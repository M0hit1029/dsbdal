```python
import pandas as pd

df = pd.read_csv('./datasets/forestfires/forestfires.csv')

print(df.head())


df_melted = df.melt(id_vars=['X','Y'],value_vars=['day','FFMC','DMC'],var_name='yo',value_name='ok')
print(df_melted)

df_casted = df_melted.pivot_table(index=['X','Y'],columns='yo',values='ok',aggfunc=list).reset_index()
print(df_casted.head())
```
